```python
#Citation for Linear Regression:
#Biswal, A. (2023, April 3). Sklearn linear regression (step-by-step explanation): Sklearn Tutorial. Simplilearn.com. 
#Retrieved from https://www.simplilearn.com/tutorials/scikit-learn-tutorial/sklearn-linear-regression-with-examples 
x_axis = np.array(linear_years).reshape(-1, 1)
y_axis = np.array(linear_cpi).reshape(-1, 1)
trainx, testx, trainy, testy = train_test_split(x_axis, y_axis, test_size = 0.9)
lin_reg = LinearRegression()
lin_reg.fit(trainx, trainy)
#print(lin_reg.score(X_test, y_test))
```

```python
data_prediction = lin_reg.predict(testx)
plt.scatter(testx, testy, color ='b')
plt.plot(testx, data_prediction, color ='k')
plt.xlabel('Year')
plt.ylabel('CPI')
plt.title('Linear Test Data 1980 - 2021')
plt.show()
```

```python
user_input = int(input("please input year after 2021: "))
```

```python
if (user_input > 2021):
    y_predict = lin_reg.intercept_ + lin_reg.coef_ * user_input
    future_predict = float(y_predict)
    print("The estimated CPI for that year is: ", round(future_predict, 2))
    #print(round(future_predict, 2))
else:
    print("Please enter a year after 2021 and try again")
```

```python
#Average CPI Over Next 5 Years
mean_years = [user_input]
mean_cpi = [y_predict]
sum_cpi = 0
for years in mean_years:
    if years < user_input + 5:
        years = years + 1
        mean_years.append(years)
        mean_cpi.append(lin_reg.intercept_ + lin_reg.coef_ * years)
        sum_cpi = sum_cpi + (lin_reg.intercept_ + lin_reg.coef_ * years)
round_cpi = float(sum_cpi / 5)    
print('The Average (Mean) CPI Over Next 5 Years From Input Date: ', round(round_cpi, 2))

#print(round(round_cpi, 2))
    
```
